```python
#drop null value
diwali.dropna(inplace=True)
```

Exploratory data analysis

gender

```python
import seaborn as sns
```

```python
#plotting a bar chart on gender and its count
ax = sns.countplot(x= 'Gender',data=diwali)
for bar in ax .containers:
                ax.bar_label(bar,color="r")
```

```python
# plotting a bar chart for gender vs total amount
sales_gen=diwali.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount',data=sales_gen,color="y")
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

AGE

```python
ax=sns.countplot(data=diwali, x='Age Group', hue='Gender')
for bars in ax.containers:
             ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
age=diwali.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=age,color="b")
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

STATE

```python
SALES_state=diwali.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
z="r"
sns.barplot(x='State',y='Orders',data=SALES_state,color=z)
# total number of orders from top 10 states
```

# total amount/sales from top 10 states


```python
sales_state = diwali.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount',color="g",edgecolor="g",alpha=1)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

marital status

```python
ax=sns.countplot(data=diwali,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
      ax.bar_label(bars) 
```

```python
sales1=diwali.groupby(['Gender','Marital_Status'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales1, x='Marital_Status',y='Amount',hue='Gender')
```

```python
#From above graphs we can see that most of the buyers are married (women) and they have high purchasing power#
```

occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=diwali,x='Occupation')
for bars in ax.containers:
 ax.bar_label(bars,color="m")
```

```python
sales_state = diwali.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
c="m"
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',color=c,edgecolor="b")
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

product category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=diwali,x='Product_Category')
for bars in ax.containers:
             ax.bar_label(bars)
```

```python
sales_state=diwali.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(19,5)})
c="y"
sns.barplot(data=sales_state, x = 'Product_Category',y='Amount',color=c,alpha=1,linestyle="-",edgecolor="b",linewidth=3)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state=diwali.groupby(['Product_ID'],as_index=False)['Orders'].count().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
c="y"
sns.barplot(data=sales_state,x='Product_ID',y='Orders',linestyle=":",edgecolor="r",linewidth=3,color=c)
```
